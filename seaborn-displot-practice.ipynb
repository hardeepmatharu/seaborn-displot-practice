### Plotting a displot, seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.displot([10, 20,20, 40],kind='hist') # displays a histogram, which shows the frequency distribution of the data by dividing into bins
plt.show()
```

Plotting a displot

Plotting a Displot without the histogram, of kind = ecdf

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.displot([0,1,2,3,4,5], kind='ecdf') #creates an empirical cumulative distribution function plot, showing the proportion of data points falling below each other.
plt.show()
```

Plotting the graph of kind = 'kde'

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.displot([1,2,3,4,5,6], kind='kde') #kernel density estimate plot, which provides smooth, continuous representation of the data's distribution.
plt.show()
```

# Difference between Normal and Binomial distribution 

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns 

data = {
    "normal": random.normal(loc=50, scale = 5, size=1000),
    "binomial": random.binomial(n=100,p=0.5, size= 1000)
}

sns.displot(data, kind="kde")
plt.show()
```

POISSON DISTRIBUTION (discrete distribution) how many times event can happen in specified time.
Parameters: 
  lam: rate or known number of occurances e.g 2 if person eats twice a day
  size: shape of returned array  

```python
from numpy import random
x = random.poisson(lam=2, size = 10)
print(x)
```

Visualization of Poisson distribution

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.displot(random.poisson(lam=2, size = 1000))
plt.show()
```

Compare normal and poisson distribution

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    "normal": random.normal(loc=50, scale=7, size=1000),
    "poisson": random.poisson(lam=50, size = 1000)
}

sns.displot(data, kind="kde")
plt.show()
```

Difference between binomial and poisson distribution
Binomial - 2 outcomes, with non zero p and large n(trials), n*p almost equal to lam

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    "binomial": random.binomial(n=1000, p =0.01, size = 1000),
    "poisson": random.poisson(lam= 10, size=1000)
}
sns.displot(data, kind = "kde")
plt.show()
```

# LOGISTIC DISTRIBUTION
    - it describes growth
    - extensively used in machine learning in logistic regression, neural networks etc.
## Parameters:
###    - loc: mean, peak , default 0
###    - scale: standard deviation, flatness of distribution, default 1
###    - size: shape of the returned array

```python
from numpy import random

x = random.logistic(loc=1, scale = 2, size=(2,3))
print(x)
```

# Visualization of Logistic Distribution

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.displot(random.logistic(size= 1000), kind="kde")
plt.show()
```

# Difference between Logistic Distribution and Normal Distribution 
##    - identical
##    - logistic distribution has more area under the tails, possibility of occurrence of an event further from mean
##    - for higher value of scale(std deviation) the normal and logistic distributions are near identical apart from the peak


```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
data = {
    "normal": random.normal(scale = 2, size = 1000),
    "logistic": random.logistic(size = 1000)
}
sns.displot(data, kind="kde")
plt.show()
```
